# Drawing Samples

Let's first generate samples from a multivariate normal distribution as follow:

```python
import numpy as np
import matplotlib.pyplot as plt

mu1 = np.array([0.0, 0.0])
mu2 = np.array([5.0, 0.0])
sigma1 = np.array([[1.0, 0.0], [0.0, 1.0]])
sigma2 = np.array([[1.0, 0.0], [0.0, 1.0]])
x1, y1 = np.random.multivariate_normal(mu1, sigma1, 1000).T
x2, y2 = np.random.multivariate_normal(mu2, sigma2, 1000).T

plt.plot(x1, y1, '.')
plt.plot(x2, y2, '.')
plt.axis('equal')
plt.show()
```

# Closed-Form 

The Kullback-Leibler divergence between two Gaussians can be obtained in closed form
$$ D_{KL}(\hat{f} \| \hat{g})  = \frac{1}{2}\left( \log\frac{det(\Sigma_{\hat g})}{det(\Sigma_{\hat f})} + Tr(\Sigma_\hat{g}^{-1}\Sigma_\hat{f}) -d + (\mu_\hat{f} - \mu_\hat{g})^\top \Sigma^{-1}_\hat{g}(\mu_\hat{f} - \mu_\hat{g}) \right) $$

For the two Gaussians above, the KL divergence should therefore be:

```python
def kl_divergence(sigma1, sigma2, mu1, mu2):
    sigma2_inv = np.linalg.inv(sigma2)
    mean_diff = (mu1 - mu2)

    return np.log(np.linalg.det(sigma1)/np.linalg.det(sigma2)) 
    + np.trace(sigma2_inv.dot(sigma1)) - mu1.shape[0] 
    + mean_diff.T.dot(sigma2_inv).dot(mean_diff)
    
kl_divergence(sigma1, sigma2, mu1, mu2)
```

# Monte-Carlo Estimation

The KL divergence between GMM unfortunately does not admit a closed form solution. We can get around this using an empirical estimate of the divergence measure:
$$ D_{MC}(f \| g) = \frac{1}{n} \sum_{i=1}^n \log \frac{f(x_i)}{g(x_i)} \to D(f \| g) $$

as $n \to \infty$. For the following, we will use the GMM implementation of scikit-learn to fit a model to a synthetic dataset. 

```python
from sklearn import mixture

nsamples = 1000
X1 = np.vstack((np.random.multivariate_normal([0.0, 0.0], [[0., -0.7], [3.5, .7]] , nsamples),
              np.random.multivariate_normal([5., 5.], [[1., 0.], [0., 1.]] , nsamples)))

X2 = np.vstack((np.random.multivariate_normal([-0.5, 5.], [[-0.4, -0.2], [6.5, .2]] , nsamples),
              np.random.multivariate_normal([-5., 0.], [[1., 0.], [0., 1.]] , nsamples)))

fig,ax = plt.subplots(1, 2)
ax[0].set_title('Model 1')
ax[1].set_title('Model 2')
ax[0].plot(X1[:, 0], X1[:, 1], '.')
ax[1].plot(X2[:, 0], X2[:, 1], '.')
```

Using Expectation-Maximization (EM), we fit a mixture model with two components to the data:

```python
gmm1 = mixture.GMM(n_components=2, covariance_type='full').fit(X1)
gmm2 = mixture.GMM(n_components=2, covariance_type='full').fit(X2)
```

In order to sample from a GMM, we would need to first sample a component according to the prior probabilities $\pi_i$ (mixing weights) and then sample from the corresponding Gaussian. In scikit-learn, this can be accomplished using [GMM.sample](http://scikit-learn.org/stable/modules/generated/sklearn.mixture.GMM.html#sklearn.mixture.GMM.sample) in conjunction with [GMM.score](http://scikit-learn.org/stable/modules/generated/sklearn.mixture.GMM.html#sklearn.mixture.GMM.score). 

```python
def kl_montecarlo(f, g, nsamples=10000):
    return np.mean(f.score(f.sample(nsamples)) - g.score(g.sample(nsamples)))

print 'The similarity property should hold (tend to 0):', kl_montecarlo(gmm1, gmm1)
print 'Positivity is not guaranteed:', kl_montecarlo(gmm1, gmm2)
```

# The Unscented Transformation

The Monte-Carlo approach presented above stems from the problem of estimating a nonlinear transformation of some random variables. Gaussian Mixture Models offer a generative way of representing and sampling from the probability distribution underlying the data points. At a first glance, it might feel odd to tackle this problem using a Kalman Filter. After all, we are not dealing with *time* nor have a state or observation model. Yet, we will see how the [Unscented Kalman Filter](http://en.wikipedia.org/wiki/Kalman_filter#Unscented_Kalman_filter) can offer a sample efficient approach for estimating the KL divergence. 

$$D_{\text{unscented}}(f \| g) = \frac{1}{2d} \sum_{i} \pi_i \sum_{k=1}^{2d} \log \frac{f(x_{i,k})}{g(x_{i,k})}$$

where a *sigma* point $x_{ik}$ for component $i$ is choosen as 
$$ x_{i,k} = \mu_i + \sqrt{d \lambda_{i,k}}U_{i,k} \\
x_{i,k+d} = \mu_i - \sqrt{d \lambda_{i,k}}U_{ik}$$

where $\lambda_{i,k}$ is the kth eigenvalue along the diagonal in the [Singular Value Decomposition (SVD) decomposition](http://en.wikipedia.org/wiki/Singular_value_decomposition) of the covariance matrix of component $i$. Similarly, $U_k$ is the kth eigenvector in the SVD.

```python
def kl_unscented(f, g):
    d = f.means_[0].shape[0]
    ncomponents = f.means_.shape[0]
    
    log_ratios = np.empty(ncomponents)
    for i, (mu, cov) in enumerate(zip(f.means_, f.covars_)):
        u, s, v = np.linalg.svd(cov)
        sigma_points = np.empty((2*d, d))
        for k in range(s.shape[0]):
            cov_sqrt = np.sqrt(d*s[k])*u[:,k]
            sigma_points[[k, k+d], :] = [mu + cov_sqrt, mu - cov_sqrt]
        print sigma_points
        plt.plot(sigma_points, '.')
        log_ratios[i] = np.sum(f.score(sigma_points) - g.score(sigma_points))
        
    return (1./(2.*d))*f.weights_.dot(log_ratios).sum()
            
plt.plot(X1[:,0], X1[:,1], '.')
kl_unscented(gmm1, gmm2)
```
